```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
from tqdm import tqdm_notebook
```

```python
## 경제-글로벌경제 뉴스 크롤링
url_base = 'https://news.naver.com/main/list.nhn'
url_pages = ['?mode=LS2D&mid=shm&sid1=101&sid2=262']
url = url_base + url_pages[0]

html = urlopen(url)
soup = BeautifulSoup(html,'html.parser')

pages= soup.find('div','paging').find_all('a')
for page in pages :
    url_pages.append(page['href'])
```

```python
# # 세계-미국/중남미 뉴스 크롤링
# url_base = 'https://news.naver.com/main/list.nhn'
# url_pages = ['?mode=LS2D&mid=shm&sid1=104&sid2=232']
# url = url_base + url_pages[0]

# html = urlopen(url)
# soup = BeautifulSoup(html,'html.parser')

# pages= soup.find('div','paging').find_all('a')
# for page in pages :
#     url_pages.append(page['href'])
```

```python
links = []
news = []
for url_page in url_pages:
    link=[]
    url = url_base + url_page
    html = urlopen(url)
    soup = BeautifulSoup(html,'html.parser')
    link.extend(soup.find_all('dt'))
    for i in range(len(link)-2):
        news.append(link[i].find('a')['href'])
    links.extend(link)
news_url = np.unique(news)
```

```python
news_url
```

```python
import re
def clean_text(text):
    cleaned_text = re.sub('[a-zA-Z]', '', text)
    cleaned_text = re.sub('[\{\}\[\]\/?.,;:|\)*~`!^\-_+<>@\#$%&\\\=\(\'\"]',
                          '', cleaned_text)
    cleaned_text = re.sub('본문 내용    플레이어     플레이어     오류를 우회하기 위한 함수 추가',
                          '', cleaned_text)
    return cleaned_text
```

```python
news_text = []
news_date = []
news_like = []
news_soso = []
news_bad = []
news_angry = []
news_want = []
# news_repl = []
for url in tqdm_notebook(news_url):
    html = urlopen(url)
    soup = BeautifulSoup(html,'html.parser')
    likit = soup.find_all('span','u_likeit_list_count _count')
    text = ''
    for item in soup.find_all('div',id = 'articleBodyContents'):
        text = text + str(item.find_all(text=True))
    news_text.append(clean_text(text))
    news_date.append(soup.find('span','t11').get_text())
    news_like.append(likit[0].get_text())
    news_soso.append(likit[1].get_text())
    news_bad.append(likit[2].get_text())
    news_angry.append(likit[3].get_text())
    news_want.append(likit[4].get_text())
#     news_repl.append(soup.find('span','u_box_count').get_text())
```

```python
data = {'Date':news_date,'Text':news_text,'Like':news_like,'Good':news_soso,'Bad':news_bad,'Angry':news_angry,'Want':news_want}
df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv('news_data.csv',sep=',',encoding='utf-8')
```
