```python
import pandas as pd
import mglearn
from sklearn.datasets import load_iris

mglearn모듈
-간단하게 글을 그리거나 필요한 데이터를 불러오기위해 사용
```

```python
iris_dataset = load_iris()
```

```python
print(iris_dataset.DESCR)
```

```python
print('Target name : {}'.format(iris_dataset['target_names']))
```

```python
print('Feature name : {}'.format(iris_dataset['feature_names']))
```

```python
print('Data type : {}'.format(type(iris_dataset['data'])))
```

```python
print('Data size : {}'.format(iris_dataset['data'].shape))
```

```python
print('First to fifth : \n\n{}'.format(iris_dataset['data'][:5]))
```

```python
print('Target type : {}'.format(type(iris_dataset['target'])))
```

```python
print('Target size : {}'.format(iris_dataset['target'].shape))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=0)
```

```python
print('X_train size : {}'.format(X_train.shape))
print('Y_train size : {}'.format(Y_train.shape))
```

```python
print('X_test size : {}'.format(X_test.shape))
print('Y_test size : {}'.format(Y_test.shape))
```

```python
iris_dataframe = pd.DataFrame(X_train,columns=iris_dataset.feature_names)
```

```python
pd.plotting.scatter_matrix(iris_dataframe, c = Y_train, figsize=(15,15), marker='o', hist_kwds={'bins':20},alpha=.8,cmap=mglearn.cm3)
```
